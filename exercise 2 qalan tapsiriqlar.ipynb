```python
#list comprehension
#exercise 10
first_letter = []
words = ["apple","banana","cherry","date"]
for x in words:
  first_letter.append(x[0])
print(first_letter)
```

```python
#exercise 11
a = []
numbers = [1,-2,3,-4,5,-6,7,-8]
for x in numbers:
  if x > 0:
    a.append(x)
print(a)
```
