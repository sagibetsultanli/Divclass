```python
#3
a = 5
b = 10
operator = 's'
if operator == '+':
  print(a+b)
elif operator == '-':
  print(a-b)
elif operator == '*':
  print(a*b)
elif operator == '/':
  print(a/b)
else :
  print('invalid operation')
```
